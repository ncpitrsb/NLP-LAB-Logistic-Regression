unigram feature

```
def featurize_uni(token_list):
    token_list = eval(token_list)
    features = {}
    for token in token_list:

        features[token] = 1
    return features
```

bigram feature

```
def featurize_bi(token_list):
    token_list = eval(token_list)
    features = {}
    for token in range(len(token_list)-1):
        features[token_list[token]+'_'+token_list[token+1]] = 1
    return features
```

uni+bi feature

```
from pythainlp.corpus import thai_stopwords
def featurize_uni_bi(token_list):
    words = thai_stopwords()
    token_list = eval(token_list)
    features = {}
    # for token in range(len(token_list)-1):
    #     features[token_list[token]+'_'+token_list[token+1]] = 1
    for token in token_list:
      if token not in words:
          features[token] = 1
    return features
```
